# 🎯 What is the Accuracy Paradox?

The **Accuracy Paradox** occurs when a classification model has **high accuracy**, but it is actually **misleading** and performs **poorly** on the real problem.

---

## 🤔 Why Does It Happen?

Because **accuracy** is the ratio of correct predictions to total predictions:

> Accuracy = (Correct Predictions) / (Total Predictions)


It **doesn't consider** class imbalance (when one class dominates the dataset).

---

## 📉 Example

Suppose you're building a model to detect a **rare disease**.

- Dataset:
  - 990 healthy people
  - 10 people with the disease

If your model **predicts everyone is healthy**, you'll get:

- ✅ Correct predictions: 990 (healthy)
- ❌ Missed disease cases: 10
- 🎯 Accuracy = 990 / 1000 = **99%**

BUT: It missed **all** disease cases = **useless model!**

---

## 💡 Better Metrics to Use Instead:

| Metric        | Why it's Better                  |
|---------------|----------------------------------|
| **Precision** | Focuses on correct positive predictions |
| **Recall**    | Focuses on catching all actual positives |
| **F1-Score**  | Harmonic mean of precision and recall |
| **Confusion Matrix** | Shows true/false positives/negatives |

---

## ✅ Summary

- **High accuracy ≠ good model**, especially with **imbalanced datasets**
- Always check **precision**, **recall**, **F1-score**, and **confusion matrix**

> 🧠 Think beyond accuracy — it can fool you!
