###  K-Means Clustering

* **Type**: Unsupervised learning
* **Goal**: Group data into **K clusters** based on similarity
* **How**:

  1. Choose K
  2. Randomly set K centroids
  3. Assign points to nearest centroid
  4. Recalculate centroids
  5. Repeat until stable
* **Used For**: Customer segmentation, image compression, etc.
* **Limitation**: Need to choose K, sensitive to starting points



###  Random Initialization Trap
What is it?
If the initial centroids are chosen poorly (e.g., too close together or in a sparse area), the algorithm may:

1. Converge to a local minimum

2. Produce suboptimal clusters

Give inconsistent results on different runs


####  Solution: K-Means++ Initialization
Instead of random centroids:

1. Choose the first centroid randomly

2. Next centroids are chosen far from existing ones

3. This leads to better spread and more reliable results
