#  Decision Tree  Simple Explanation

##  What is a Decision Tree?

A **Decision Tree** is a popular machine learning algorithm used for:
- **Classification** (e.g., Yes or No)
- **Regression** (predicting numbers)

It works like a **flowchart**:
- Each **node** represents a feature (question)
- Each **branch** represents a decision (answer)
- Each **leaf node** represents a final output (class/label)

---

##  How Does it Work?

1. The tree starts at the **root node**.
2. It checks one feature and splits the data based on the answer.
3. This process continues (like asking questions) until:
   - A final decision is reached
   - Or no further splits are useful

---

##  Example: Should You Play Outside?

Suppose you want to decide if you should play outside based on weather:

| Weather | Temperature | Play? |
|---------|-------------|--------|
| Sunny   | Hot         | No     |
| Sunny   | Mild        | Yes    |
| Rainy   | Cool        | No     |
| Overcast| Mild        | Yes    |
| Sunny   | Cool        | Yes    |
