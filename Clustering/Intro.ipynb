### 🆚 Classification vs Clustering

| Feature                | **Classification**                                | **Clustering**                                     |
| ---------------------- | ------------------------------------------------- | -------------------------------------------------- |
| **Type**               | Supervised Learning                               | Unsupervised Learning                              |
| **Labeled Data**       | ✅ Yes – Uses labeled data                         | ❌ No – Works with unlabeled data                   |
| **Goal**               | Predict predefined labels or classes              | Group similar data points into clusters            |
| **Output**             | Class/Category (e.g., Spam / Not Spam)            | Cluster ID (e.g., Cluster 1, Cluster 2...)         |
| **Examples**           | - Email spam detection<br>- Disease diagnosis     | - Customer segmentation<br>- Market research       |
| **Algorithms Used**    | - Logistic Regression<br>- Decision Tree<br>- SVM | - K-Means<br>- DBSCAN<br>- Hierarchical Clustering |
| **Evaluation Metrics** | Accuracy, Precision, Recall, F1 Score             | Silhouette Score, Davies-Bouldin Index             |

---

### 🧠 Example to Understand Better

#### 🔷 Classification:

> You train a model on emails labeled as "Spam" or "Not Spam". The model learns from this and predicts future emails into those categories.

#### 🔶 Clustering:

> You have a bunch of customer data but no labels. Clustering will group similar customers together (e.g., based on age, spending) without knowing the labels in advance.
